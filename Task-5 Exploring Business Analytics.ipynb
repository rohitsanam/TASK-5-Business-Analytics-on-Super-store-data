## Let us have a look at the profit distribution with respect to shipmode

```python
plt.figure(figsize=(15,8))
sns.barplot(x=superstore['Ship Mode'],y=superstore['Profit'])
```

- **From the above barplot we can see that the maximum profit is done by the superstore when the shipment was delivered or shipped the same day followed by First class,Second class,Standard class**

## Let's have a look at the profit with respect to segment

```python
plt.figure(figsize=(15,8))
sns.barplot(x=superstore['Segment'],y=superstore['Profit'])
```

- **From the above barplot we can see that the maximum profit was made by the Home Office segment followed by Corporate segment and then followed by Consumer**

## Let's Have a look at the profit with respect to State

```python
top5_state_ptrofit = superstore.groupby('State').agg({'Profit':'max'}).head(5)
top5_state_ptrofit
```

```python
states__with_less_profits = superstore.groupby('State').agg({'Profit':'min'}).tail(5)
states__with_less_profits
```

```python
plt.figure(figsize=(20,21))
sns.barplot(x=superstore['Profit'],y=superstore['State'])
```

- **From the above barplot and the two dataframes for the states with less profits and maximum profits we can say that Albama is the most profitable state and West Virginia is least profitable state**

## Let's have a look at the profit with respect to the region.

```python
plt.figure(figsize=(15,8))
sns.barplot(x=superstore['Region'],y=superstore['Profit'])
```

- **From the above barplot we can see that the most profitable region is the West Region and the least profitable region is the Central Region**

## Lets have a look at the profit category-wise.

```python
plt.figure(figsize=(15,8))
sns.barplot(x=superstore['Category'],y=superstore['Profit'])
```

## Lets look at the relationship between sales and profit

```python
plt.figure(figsize=(15,8))
plt.scatter(x=superstore['Sales'],y=superstore['Profit'],)
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.xlim(0,20000)
plt.ylim(0,8000)
plt.title('Sales V/s Profit')
```

- **From the above scatter plot we can see that there is a strong positive correlation between Sales and Profit**

## Let's have a look at the relationship between Quantity and profit

```python
plt.figure(figsize=(15,8))
plt.scatter(x=superstore['Quantity'],y=superstore['Profit'])
plt.xlabel('Quantity')
plt.ylabel('Profit')
plt.ylim(0,8000)
plt.title('Quantity v/s Profit')
```

- **There is no linear relationship and no correlation between Quantity and profit**
